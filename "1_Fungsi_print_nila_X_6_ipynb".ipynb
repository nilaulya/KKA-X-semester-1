**Parameter Fungsi Print**

```python
tahun=2000
bulan= "Februari"
tanggal=18

print(tahun, end="-")
print(bulan, end="-")
print(tanggal)
```

```python
print(tanggal, end="/")
print(bulan, end="/")
print(tahun)
```
